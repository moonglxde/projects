```python
import pandas as pd
url = 'https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv'
df = pd.read_csv(url)
print("Original shape:", df.shape)
```

```python
df_clean = df.dropna()  # Drop nulls
df_filtered = df_clean[df_clean['total_bill'] > 10]  # Filter
grouped = df_filtered.groupby('day')['total_bill'].sum()
print(grouped)
df_clean.to_csv('clean_tips.csv', index=False)
```
