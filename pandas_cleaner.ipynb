```python
import pandas as pd
url = 'https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv'
df = pd.read_csv(url)
print("Original shape:", df.shape)
```

```python
df_clean = df.dropna()  # Drop nulls
df_filtered = df_clean[df_clean['total_bill'] > 10]  # Filter
grouped = df_filtered.groupby('day')['total_bill'].sum()
print(grouped)
df_clean.to_csv('clean_tips.csv', index=False)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.plot(df['total_bill'], label='Bill Amount', color='blue', linewidth=2)
plt.title('Restaurant Bill Trends', fontsize=16, fontweight='bold')
plt.xlabel('Customer Index')
plt.ylabel('Total Bill ($)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
day_bills = df.groupby('day')['total_bill'].mean()
plt.figure(figsize=(8, 5))
day_bills.plot(kind='bar', color=['skyblue', 'lightcoral', 'gold', 'lightgreen'])
plt.title('Average Bill by Day of Week', fontsize=16, fontweight='bold')
plt.xlabel('Day')
plt.ylabel('Average Bill ($)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
print("Average bills by day:\n", day_bills)
```
