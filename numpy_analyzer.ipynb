# NumPy Array Analyzer
Simulates ESP32 sensor data. Skills: array operations, statistics.


```python
import numpy as np
data = np.random.randn(100)  # Random sensor-like data
print("Mean:", np.mean(data))
print("Std:", np.std(data))
reshaped = data.reshape(10, 10)
print("Reshaped shape:", reshaped.shape)
```

```python
np.savetxt('sensor_data.csv', data, delimiter=',')
print("✓ Exported sensor_data.csv - 100 sensor readings saved!")
```
