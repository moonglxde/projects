```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
print("ðŸš€ ALL LIBRARIES LOADED!")
```

```python
# Project 1: IoT Sensor Data (NumPy)
sensor_data = np.random.normal(25, 5, 100)  # 100 temp readings
iot_df = pd.DataFrame({'temperature': sensor_data, 'sensor_id': range(1,101)})

# Project 2: Restaurant Tips (Pandas)
tips_url = 'https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv'
tips_df = pd.read_csv(tips_url).dropna()

# COMBINED Dataset
print("IoT Sensors:", iot_df.shape)
print("Restaurant Tips:", tips_df.shape)
print("\nâœ… Data from ALL projects loaded!")
```

```python
# Restaurant insights (Project 2 enhanced)
day_summary = tips_df.groupby(['day', 'time'])['total_bill'].agg(['mean', 'count']).round(2)
print("Daily Bill Patterns:\n", day_summary)

# Sensor insights (Project 1 enhanced)  
iot_summary = iot_df.groupby(pd.cut(iot_df['temperature'], bins=5), observed=False).size()
print("\nTemperature Distribution:\n", iot_summary)
```

```python
# Advanced NumPy analysis
print("ðŸ”¬ SENSOR STATISTICS:")
print(f"Mean: {sensor_data.mean():.2f}Â°C")
print(f"Std: {sensor_data.std():.2f}Â°C") 
print(f"Outliers: {(sensor_data > 30).sum()} high temps")

# Correlation matrix
corr_matrix = np.corrcoef(sensor_data, range(100))
print(f"Trend correlation: {corr_matrix[0,1]:.3f}")
```

```python
plt.style.use('dark_background')
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# 1. IoT Temp Distribution (NumPy â†’ Plot)
axes[0,0].hist(sensor_data, bins=20, color='cyan', alpha=0.7, edgecolor='white')
axes[0,0].set_title('IoT Temperature Distribution', fontsize=14, fontweight='bold')

# 2. Bill vs Tip (Pandas â†’ Seaborn)
sns.scatterplot(data=tips_df, x='total_bill', y='tip', hue='day', ax=axes[0,1])
axes[0,1].set_title('Bill vs Tip by Day')

# 3. Daily Bill Trends (Pandas groupby â†’ Plot)
day_bills = tips_df.groupby('day')['total_bill'].mean()
axes[1,0].bar(day_bills.index, day_bills.values, color='gold')
axes[1,0].set_title('Avg Bill by Day')

# 4. Sensor Time Trend (NumPy â†’ Line)
axes[1,1].plot(sensor_data[:50], color='red', linewidth=2)
axes[1,1].set_title('First 50 Sensor Readings')

plt.tight_layout()
plt.show()
```

```python
# Interactive restaurant dashboard
fig = px.scatter(tips_df, x='total_bill', y='tip', 
                size='size', color='day', hover_data=['sex', 'smoker'],
                title="Interactive Restaurant Analytics")
fig.show()
```

```python
# Export complete analysis
iot_df.to_csv('complete_pipeline_data.csv', index=False)
day_summary.to_csv('restaurant_insights.csv')

print("ðŸŽ‰ PIPELINE COMPLETE!")
print("\nðŸ“Š Portfolio Files Created:")
print("- complete_pipeline_data.csv (IoT + Restaurant)")
print("- restaurant_insights.csv (Business insights)")
print("\nâœ… Skills: NumPy, Pandas, Matplotlib, Seaborn, Plotly")
print("\nðŸŒ LIVE: github.com/moonglxde/projects")
```
